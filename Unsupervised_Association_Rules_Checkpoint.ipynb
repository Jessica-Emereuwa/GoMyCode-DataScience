```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Provided toy dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
# Transform the dataset into a format suitable for the apriori algorithm
te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset) #Apply one-hot-encoding on our dataset
df = pd.DataFrame(te_ary, columns=te.columns_) #Creating a nw Dataframe from our numpy array

df
```

```python
#Try to explore the checkpoint dataset using Pandas and Plotly.
import plotly.express as px

# Calculate item frequencies
item_frequencies = df.sum()  # Calculate sum of each column (item)

# Create a DataFrame for plotting
df_plot = pd.DataFrame({'Item': item_frequencies.index, 'Frequency': item_frequencies.values})

# Bar plot of item frequencies
fig = px.bar(df_plot, x='Item', y='Frequency', title='Frequency of Items in Toy Dataset')
fig.show()
```

```python
# Apply the apriori algorithm
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
frequent_itemsets
```

**INTERPRET THE RESULT**



*  Pants and scarves are the most frequently purchased items, appearing in 80% of transactions.

*   The combination of hat, sneakers, pants, and skirt appears in 40% of transactions.


*   Pants and scarf are frequently bought together, appearing in 60% of transactions.

```python
# Generate the association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

```python
print("Frequent Itemsets:\n", frequent_itemsets)
```

```python
print("\nAssociation Rules:\n", rules)
```

**Interpretation of Association Rules**



*   Dress and Heels: Customers who buy dresses also buy heels with 100% confidence, indicating a strong correlation (lift of 2.5).


*   Hat, Sneakers, Pants, and Skirt: The combination of these items frequently occurs together, indicating a strong relationship (lift of 2.5).


*   Dress and Scarf: Dresses are often purchased with scarves, showing a confidence of 100% but with a lower lift (1.25).


*   Earrings and Scarf: Earrings are frequently bought with scarves, also with a confidence of 100% and a lift of 1.25.





**Business Plan Recommendations**


*   Create promotional bundles for dresses and heels. Given the strong association, customers are likely to appreciate discounts or special offers on these combinations.

*   Since dresses are frequently purchased with scarves, consider offering a discount on scarves with the purchase of a dress. This can encourage customers to buy both items.
