```python
"""
Instructions

Load the dataset, display the ten first lines, store the results in a variable called 'client_0_bills'.
What is the data type of the 'client_0_bills' variable ?
Display the general information of the dataset and try to answer the following questions :
How many rows and columns do we have in this dataset ?
How many categorical features are present in the dataset ?
How much memory space does the dataset consume ?
Inspect the dataset for potential missing values.
Select your strategy to handle missing values, and tell us why you had made that choice.
Run a descriptive analysis on numeric features (columns).
Select the bills records for the client with an id ='train_Client_0', using 2 methods.
Transform the 'counter_type' feature to a numeric variable using the encoder of your choice.
Delete the 'counter_statue' feature from the Dataframe

How We'll Evaluate

Your checkpoint will be evaluated based on these criteria, each rated from 0 to 5.

What We're Looking For

Successfully import the dataset using Pandas.

Use appropriate methods to identify missing data

Successfully transform categorical data into a numerical representation

Successfully apply filters on selection

Successfully delete columns
"""


import pandas as pd

# Load the dataset and display the first ten lines
client_0_bills = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/STEG_BILLING_HISTORY.csv')
client_0_bills.head(10)
```

```python
# Check the data type of the 'client_0_bills' variable
type(client_0_bills)
```

```python
# Display general information of the dataset
client_0_bills.info()
```

```python
# Count rows and columns
rows, cols = client_0_bills.shape
print("Number of rows:", rows)
print("Number of columns:", cols)
```

```python
# Count categorical features
cat_features = client_0_bills.select_dtypes(include=['object']).columns
num_cat_features = len(cat_features)
print("Number of categorical features:", num_cat_features)
```

```python
# Memory consumption of the dataset
memory_consumption = client_0_bills.memory_usage().sum() / (1024*1024) # in MB
print("Memory consumption:", memory_consumption, "MB")
```

```python
# Check for missing values
#this sums the missing values in each columns

missing_values = client_0_bills.isnull().sum()
print("Missing values:")
missing_values
```

```python
# Handling missing values strategy:
#Mean is used to replace missing values because the missing data is numerical

# Impute missing values for numeric features with mean
numeric_features = client_0_bills.select_dtypes(include='number').columns
client_0_bills[numeric_features] = client_0_bills[numeric_features].fillna(client_0_bills[numeric_features].mean())
client_0_bills
```

```python

# Descriptive analysis of numeric features
print(client_0_bills.describe(include='number'))
```

```python
# Select bills records for the client with id ='train_Client_0' using two methods
client_0_bills_method1 = client_0_bills[client_0_bills['client_id'] == 'train_Client_0']
client_0_bills_method2 = client_0_bills.query("client_id == 'train_Client_0'")

```

```python
# Perform one-hot encoding on the 'counter_type' feature
client_0_bills = pd.get_dummies(client_0_bills, columns=['counter_type'], prefix='counter_type')

# Display the updated DataFrame
client_0_bills.head()
```
