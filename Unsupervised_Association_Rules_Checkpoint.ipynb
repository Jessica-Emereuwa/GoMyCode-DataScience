```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Provided toy dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
# Transform the dataset into a format suitable for the apriori algorithm
te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset) #Apply one-hot-encoding on our dataset
df = pd.DataFrame(te_ary, columns=te.columns_) #Creating a nw Dataframe from our numpy array

df
```

```python
#Try to explore the checkpoint dataset using Pandas and Plotly.
import plotly.express as px

# Calculate item frequencies
item_frequencies = df.sum()  # Calculate sum of each column (item)

# Create a DataFrame for plotting
df_plot = pd.DataFrame({'Item': item_frequencies.index, 'Frequency': item_frequencies.values})

# Bar plot of item frequencies
fig = px.bar(df_plot, x='Item', y='Frequency', title='Frequency of Items in Toy Dataset')
fig.show()
```

```python
# Apply the apriori algorithm
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
frequent_itemsets
```

**INTERPRET THE RESULT**



*  Pants and scarves are the most frequently purchased items, appearing in 80% of transactions.

*   The combination of hat, sneakers, pants, and skirt appears in 40% of transactions.


*   Pants and scarf are frequently bought together, appearing in 60% of transactions.

```python
# Generate the association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

```python
print("Frequent Itemsets:\n", frequent_itemsets)
```

```python
print("\nAssociation Rules:\n", rules)
```

**Interpretation of Association Rules**



*   Dress and Heels: Customers who buy dresses also buy heels with 100% confidence, indicating a strong correlation (lift of 2.5).


*   Hat, Sneakers, Pants, and Skirt: The combination of these items frequently occurs together, indicating a strong relationship (lift of 2.5).


*   Dress and Scarf: Dresses are often purchased with scarves, showing a confidence of 100% but with a lower lift (1.25).


*   Earrings and Scarf: Earrings are frequently bought with scarves, also with a confidence of 100% and a lift of 1.25.





**Business Plan Recommendations**


*   Create promotional bundles for dresses and heels. Given the strong association, customers are likely to appreciate discounts or special offers on these combinations.

*   Since dresses are frequently purchased with scarves, consider offering a discount on scarves with the purchase of a dress. This can encourage customers to buy both items.





```python
import pandas as pd

# Assuming the dataset is in a CSV file named 'customer_purchases.csv'
df = pd.read_csv('Market_Basket_Optimisation.csv', header =None)

# Display the first few rows
df.head()
```

```python
import pandas as pd
import plotly.express as plt

Items = {}
# Access the first column (index 0) since there are no headers
for item in df.iloc[:, 0]:
    if item in Items:
        Items[item] = Items[item] + 1
    else:
        Items[item] = 1

keys = []
vals = []
for i, k in Items.items():
    if k > 30:
        keys.append(i)
        vals.append(k)

```

```python
# Create a DataFrame for Plotly Express
df_plot = pd.DataFrame({'Items': keys, 'Transactions': vals})

# Use the 'text' argument to display labels on the bars
fig = plt.bar(df_plot, x='Items', y='Transactions', text='Transactions')

fig.update_layout(
    xaxis_title='Items Sold',
    yaxis_title='Number of Transactions',
    title='Items frequency',
    xaxis_tickangle=90,
    width=1000,  # Adjust width as needed
    height=600   # Adjust height as needed
)

fig.show()
```

```python
# Display general information about the dataset
df.info()
```

```python
import numpy as np

# Convert DataFrame to list of lists
transactions = df.applymap(str).values.tolist()

# Remove 'nan' values from the transactions
transactions = [[item for item in transaction if item != 'nan'] for transaction in transactions]

# Display the first few transactions
transactions[:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Convert dataset into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_onehot = pd.DataFrame(te_ary, columns=te.columns_)

df_onehot.head()
```

```python
# Run Apriori algorithm
frequent_itemsets = apriori(df_onehot, min_support=0.01, use_colnames=True)  # Adjust min_support as needed

frequent_itemsets.head()
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

print("Frequent Itemsets:\n")
frequent_itemsets.head()

```

```python
print("\nAssociation Rules:\n")
rules.head()
```

### Interpretation of the Results


#### Association Rules:

1. **Rule 1**: (eggs, ground beef)  (mineral water)
   - **Antecedent Support**: 0.019997 (approx. 2% of transactions include both eggs and ground beef)
   - **Consequent Support**: 0.238368 (mineral water appears in approx. 23.84% of transactions)
   - **Support**: 0.010132 (approx. 1.01% of transactions include eggs, ground beef, and mineral water)
   - **Confidence**: 0.506667 (50.67% of transactions with eggs and ground beef also include mineral water)
   - **Lift**: 2.125563 (mineral water is 2.13 times more likely to be bought with eggs and ground beef than by itself)

2. **Rule 2**: (milk, ground beef)  (mineral water)
   - **Antecedent Support**: 0.021997 (approx. 2.20% of transactions include both milk and ground beef)
   - **Consequent Support**: 0.238368 (mineral water appears in approx. 23.84% of transactions)
   - **Support**: 0.011065 (approx. 1.11% of transactions include milk, ground beef, and mineral water)
   - **Confidence**: 0.503030 (50.30% of transactions with milk and ground beef also include mineral water)
   - **Lift**: 2.110308 (mineral water is 2.11 times more likely to be bought with milk and ground beef than by itself)
